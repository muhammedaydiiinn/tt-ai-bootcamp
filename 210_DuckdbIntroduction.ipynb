```python
%%duckdb

select *, sum(pc) over(partition by VendorID, pickup_date order by tpep_pickup_datetime) as cum_pc
from
(select VendorID,cast(tpep_pickup_datetime as date) as pickup_date,tpep_pickup_datetime,passenger_count as pc 
from 's3://tt-bootcamp-shared/nyc-taxi-trip/yellow-taxi-data/*.parquet' )
limit 10
```

```python
%%duckdb


select time, type, avg(has_ai) over(order by time rows between 100 preceding and 100 following) ma_ai_has from
(select time, type, cast( (case when text is null then '' else text end like '% AI %') as int) as has_ai, text  
from 's3://tt-bootcamp-shared/hackernews.parquet' )
order by 1 desc
limit 10
```

```python
%%duckdb


select time, type, avg(has_ai) over(order by time range between INTERVAL 180 DAYS preceding and INTERVAL 180 DAYS following) ma_ai_has from
(select time, type, cast( (case when text is null then '' else text end like '% AI %') as int) as has_ai, text  
from 's3://tt-bootcamp-shared/hackernews.parquet' )
order by 1 desc
limit 10
```

```python
%%duckdb

select * from 
(select *, lead(ss,1) over(partition by bikeid order by starttime) as next_trip_ss,
lead(starttime,1) over(partition by bikeid order by starttime) as next_trip_starttime
from 
(select bikeid, starttime,"start station id" as ss, "end station id" es 
from 's3://tt-bootcamp-shared/nyc-bike-trip/201601.parquet') )  where next_trip_ss != es order by 1
```

```python
%%duckdb

select  d1.Zone as pu_zone, d2.Zone as do_zone, avg(total_amount) ,max(total_amount), median(total_amount), count(1) as n
from 's3://tt-bootcamp-shared/nyc-taxi-trip/green-taxi-data/*.parquet' f
JOIN 's3://tt-bootcamp-shared/nyc-taxi-trip/zone-lookup/taxi_zone_lookup.parquet' d1 on(f.PULocationID = d1.LocationID)
JOIN 's3://tt-bootcamp-shared/nyc-taxi-trip/zone-lookup/taxi_zone_lookup.parquet' d2 on(f.DOLocationID = d2.LocationID)
group by 1,2
having count(1) > 10
order by 3 desc
limit 10
```
